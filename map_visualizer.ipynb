````python
import os
from dotenv import load_dotenv
from arcgis.gis import GIS

load_dotenv()
````

### VisualizaciÃ³n de mapas

````python
gis = GIS(api_key = os.getenv("GIS_API_KEY"))

# Creamos el widget de mapa
map = gis.map()
map.basemap.basemap = "arcgis-streets"

# Setemos el widget en el centro del mapa y agregamos el control del zoom
map.center = [18.486058, -69.931211]
map.zoom = 12

# Mostramos el mapa
map
````
