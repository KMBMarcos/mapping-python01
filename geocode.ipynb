# Visualizacion de Mapas y Esquemas

### Importamos dependencias

```python
import os
from dotenv import load_dotenv
from arcgis.gis import GIS
from arcgis.geocoding import geocode, Geocoder

load_dotenv()
```

### Geocode

```python
gis = GIS(api_key = os.getenv("GIS_API_KEY"))

geocoder_url= Geocoder("https://geocode-api.arcgis.com/arcgis/rest/services/World/GeocodeServer")


geocoder_result = geocode(address="123 Main St, Anytown, USA", geocoder=geocoder_url)

geocoder_result
```
